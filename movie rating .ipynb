````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python

df = pd.read_csv("IMDb Movies India.csv", encoding='latin1')
df.head()
````

````python
df.info()
````

````python
df.isnull().sum()
````

````python
df = df[['Genre', 'Director', 'Actor 1','Actor 2', 'Actor 3', 'Rating']]
````

````python
df.dropna(inplace=True)
````

````python
df.head()
````

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Genre'] = le.fit_transform(df['Genre'])

df['Director'] = le.fit_transform(df['Director'])

df['Actor 1'] = le.fit_transform(df['Actor 1'])

df['Actor 2'] = le.fit_transform(df['Actor 2'])

df['Actor 3'] = le.fit_transform(df['Actor 3'])
````

````python

x = df.drop('Rating', axis=1)
y = df['Rating']
````

````python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42
)
````

````python
from sklearn.ensemble import RandomForestRegressor
````

````python
model = RandomForestRegressor()
````

````python
model.fit(x_train, y_train)
````

````python
predictions = model.predict(x_test)

print(predictions)
````

````python
score = model.score(x_test, y_test)

print("Model Score:", score)
````

````python
plt.figure(figsize=(8,5))

plt.scatter(y_test,predictions)

plt.xlabel("Actual Ratings")

plt.ylabel("Predicted Ratings")

plt.title("Actual Vs Predicted Movie Ratings")

plt.show()
````

````python
sample = pd.DataFrame(
    [[1,1,1,1,1]],
    columns=x.columns
)
result = model.predict(sample)

print("Predicted Movie Ratings:",result[0])
````

````python
print("Movie Rating Prediction Project Completed Succesfully")

print("Final Model Score:",score)
````
